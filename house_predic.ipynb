```python
import pandas as pd
data_train = pd.read_csv('train.csv')
data_train.head()
```

```python
Y = data_train['SalePrice'].values
X = pd.DataFrame(data_train, columns = ['OverallQual','TotalBsmtSF'])
```

```python
from sklearn.neural_network import MLPRegressor
clf = MLPRegressor(random_state=1)
clf.fit(X, Y)
```

```python
test = pd.read_csv('test.csv')
test_X = pd.DataFrame(test, columns = ['OverallQual','TotalBsmtSF'])
test_X['TotalBsmtSF'].fillna(test_X['TotalBsmtSF'].mode().iloc[0], inplace=True)
result = clf.predict(test_X)
```

```python
samplesubmission = pd.read_csv('sample_submission.csv')
output = pd.DataFrame({'Id': samplesubmission.Id, 'SalePrice': result})
output.to_csv('submission.csv', index=False)
```
